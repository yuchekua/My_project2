```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('auto-mpg.csv', header=None)
```

```python
df
```

```python
y = df[0]
df = df.drop(columns=[0])
df, y
```

```python
x_train, x_test, y_train, y_test = train_test_split(df,y,test_size=0.5, random_state=0)
model = linear_model.LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

row = np.append(model.intercept_, model.coef_)
row = np.append(row, mean_squared_error(y_test,y_pred))
row = np.append(row, model.score(x_test,y_test))
cols = ['intercept', 'coef_1', 'coef_2','coef_3', 'coef_4', 'coef_5','coef_6', 'coef_7', 'MSE', 'R2']
results = pd.DataFrame([row], columns=  cols, index = ['MLR'])
results
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df)
df2 = pd.DataFrame(scaled)
df2
```

```python
x_train, x_test, y_train, y_test = train_test_split(df2,y,test_size=0.5, random_state=0)
model1 = linear_model.LinearRegression()
model1.fit(x_train,y_train)
y_pred1 = model1.predict(x_test)

row = np.append(model1.intercept_, model1.coef_)
row = np.append(row, mean_squared_error(y_test,y_pred1))
row = np.append(row, model1.score(x_test,y_test))
MLR_SS_results = pd.DataFrame([row], columns = cols, index=['MLR_SS'])
results = pd.concat([results, MLR_SS_results]).round(3)
results
```
