```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([[9,16],[17,4],[7,0],[9,6],[21,4],[15,12],[22,6],[13,14]])
```

```python
plt.scatter(points[:, 0], points[:, 1], s=50, cmap='viridis')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
DBS = DBSCAN(eps=4, min_samples=2).fit(points)	
y_cls = DBS.labels_
```

```python
## core points
core_points = points[DBS.core_sample_indices_]
print(f'core points: {core_points}')
```

```python
## no border points
```

```python
## outliers
Noisy = points[y_cls == -1]
print(f'Noisy points: {Noisy}')
```

```python
## plot
plt.scatter(points[:,0],points[:,1],c=y_cls)
## three clusters including noisy
```

```python
# 4b matrix
distances_matrix = np.array([np.linalg.norm((item*np.ones((len(points),len(item))))-points,axis=1) for item in points])
print(distances_matrix.round())
```
