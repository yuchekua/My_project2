```python
X = data_CSV.values
X.shape
```

```python
from sklearn.decomposition import PCA
pca_13 = PCA(n_components=13, random_state=40)
pca_13.fit(X)
x_pca_13 = pca_13.transform(X)

print("variance explained by the First Principle component = ", 
np.cumsum(pca_13.explained_variance_ratio_ * 100)[0])

print("variance explained by the First 2 Principle component = ", 
np.cumsum(pca_13.explained_variance_ratio_ * 100)[1])

print("variance explained by the First 3 Principle component = ", 
np.cumsum(pca_13.explained_variance_ratio_ * 100)[2])
```

```python
pca_13.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.scatterplot(x=x_pca_13[:,0], y=x_pca_13[:,1], s= 70,hue=data_sk.target,
palette=['green','blue','purple'])

plt.xlabel('First Principle component')
plt.ylabel('Second Principle component')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
scaled_pca_13 = PCA(n_components=13, random_state=40)
scaled_pca_13.fit(X_scaled)
Xscaled_pca_13 = scaled_pca_13.transform(X_scaled)

print("variance explained by the First Principle component = ", 
np.cumsum(scaled_pca_13.explained_variance_ratio_ * 100)[0])

print("variance explained by the First 2 Principle component = ", 
np.cumsum(scaled_pca_13.explained_variance_ratio_ * 100)[1])

print("variance explained by the First 3 Principle component = ", 
np.cumsum(scaled_pca_13.explained_variance_ratio_ * 100)[2])
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=Xscaled_pca_13[:,0], y=Xscaled_pca_13[:,1], s= 70,hue=data_sk.target,
palette=['green','blue','purple'])

plt.xlabel('First Principle component')
plt.ylabel('Second Principle component')
```
